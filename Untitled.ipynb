## **Разбираемся с кластерами**

**Чекаем количество кластеров режимов машины**

```python
# Выбор признаков, характеризующих поведение
features = [
    'Обороты ДВС',
    'Скорость',
    'Масса груза',
    'Нагрузка ДВС',
    'Напряжение сети',
    'Угол кузова (лайт)'
]

# Оставляем только нужные признаки
df_filtered = df[features].copy()

# Преобразуем строки с запятыми в числа
for col in df_filtered.columns:
    df_filtered[col] = (
        df_filtered[col]
        .astype(str)
        .str.replace(',', '.', regex=False)
        .replace('-', np.nan)
        .replace('nan', np.nan)
    )

# Преобразуем всё в float
df_filtered = df_filtered.astype(float)

# Заполняем пропуски средними значениями по каждому признаку
df_filtered.fillna(df_filtered.mean(), inplace=True)

# Масштабирование
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_filtered)

# Метод локтя и silhouette score
inertia = []
silhouette = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)
    silhouette.append(silhouette_score(scaled_data, kmeans.labels_))

# Построение графиков
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(k_range, inertia, 'bo-')
plt.xlabel('Количество кластеров')
plt.ylabel('Inertia')
plt.title('Метод локтя')

plt.subplot(1, 2, 2)
plt.plot(k_range, silhouette, 'go-')
plt.xlabel('Количество кластеров')
plt.ylabel('Silhouette Score')
plt.title('Silhouette-оценка')

plt.tight_layout()
plt.show()
```

✅ Шаг 3: Подготовка данных по каждому кластеру

```python
cluster_datasets = {}

for cluster_id in range(8):  # k = 8
    cluster_data = df_full[df_full['Cluster'] == cluster_id].copy()

    # Удалим нечисловые и ненужные поля (номер кластера уже не нужен)
    cluster_features = cluster_data.select_dtypes(include=['float64', 'int64']).drop(columns=['Cluster'], errors='ignore')

    # Масштабируем отдельно для каждого кластера
    scaler = StandardScaler()
    cluster_scaled = scaler.fit_transform(cluster_features)

    cluster_datasets[cluster_id] = {
        'X': cluster_scaled,
        'scaler': scaler
    }
```

✅ Шаг 4: Обучение AutoEncoder для каждого режима (тут начинается жара)

```python
from tensorflow.keras import backend as K

def adaptive_loss(y_true, y_pred):
    error = y_true - y_pred
    abs_error = K.abs(error)
    
    # Плавное взвешивание (логарифмическое)
    weights = 1 / (K.log(abs_error + 10) + 0.5)  
    
    # Комбинация MSE и MAE
    return K.mean(weights * (0.3*K.square(error) + 0.7*abs_error))

save_dir = 'saved_models'
os.makedirs(save_dir, exist_ok=True)

autoencoders = {}

for cluster_id, data in cluster_datasets.items():
    X = data['X']
    scaler = data['scaler']
    input_dim = X.shape[1]

    # Архитектура модели
    input_layer = Input(shape=(input_dim,))
    encoded = Dense(16, activation='relu')(input_layer)
    encoded = Dense(8, activation='relu')(encoded)
    decoded = Dense(16, activation='relu')(encoded)
    output_layer = Dense(input_dim, activation='linear')(decoded)

    autoencoder = Model(inputs=input_layer, outputs=output_layer)
    
    # Используем взвешенную функцию потерь
    autoencoder.compile(optimizer='adam', loss=adaptive_loss)

    # Обучение с callback'ами для мониторинга
    history = autoencoder.fit(
        X, X, 
        epochs=50, 
        batch_size=64, 
        shuffle=True, 
        verbose=1,
        validation_split=0.1  # Добавляем валидацию
    )

    # Сохраняем модель
    model_path = os.path.join(save_dir, f'autoencoder_cluster_{cluster_id}.h5')
    autoencoder.save(model_path)

    # Сохраняем scaler
    scaler_path = os.path.join(save_dir, f'scaler_cluster_{cluster_id}.pkl')
    joblib.dump(scaler, scaler_path)

    autoencoders[cluster_id] = {
        'model': autoencoder,
        'scaler': scaler,
        'features': list(cluster_features.columns),
        'history': history.history  # Сохраняем историю обучения
    }
```

Тут пытаемся визуализиров, определяем топ 3 признака по важности и смотрим как соответствует им то, что на восстанавливал автоэнкодер

```python
for cluster_id, cluster_data in cluster_datasets.items():
    # Получаем модель, данные и scaler для текущего кластера
    autoencoder = autoencoders[cluster_id]['model']
    X = cluster_data['X']
    scaler = cluster_data['scaler']
    features = autoencoders[cluster_id]['features']  # Названия признаков
    
    # Реконструкция данных
    reconstructed = autoencoder.predict(X)
    X_original = scaler.inverse_transform(X)
    reconstructed_original = scaler.inverse_transform(reconstructed)
    
    # --- Анализ важности признаков ---
    # Вариант 1: Через веса первого слоя (input -> encoded)
    weights = autoencoder.layers[1].get_weights()[0]  # shape = (input_dim, 16)
    importance = np.sum(np.abs(weights), axis=1)  # Сумма абсолютных весов
    
    # Вариант 2: Через MSE по признакам (раскомментировать при необходимости)
    #mse_by_feature = np.mean(np.power(X_original - reconstructed_original, 2), axis=0)
    #importance = mse_by_feature
    
    # Топ-3 важных признаков
    top_features_idx = np.argsort(importance)[-3:][::-1]
    
    # --- Визуализация ---
    print(f"\nКластер {cluster_id} | Топ-3 признаков:")
    for i, feature_idx in enumerate(top_features_idx, 1):
        feature_name = features[feature_idx]
        print(f"{i}. {feature_name} (важность: {importance[feature_idx]:.2f})")
        
        plt.figure(figsize=(10, 4))
        plt.plot(X_original[:, feature_idx], label='Исходные', alpha=0.7)
        plt.plot(reconstructed_original[:, feature_idx], 
                label='Реконструированные', alpha=0.7, linestyle='--')
        plt.title(f'Кластер {cluster_id} | Признак: {feature_name}\nВажность: {importance[feature_idx]:.2f}')
        plt.xlabel('Номер примера')
        plt.ylabel('Значение')
        plt.legend()
        plt.grid()
        plt.show()
    
    # Дополнительно: Bar-plot важности всех признаков
    plt.figure(figsize=(12, 4))
    plt.bar(range(len(importance)), importance, tick_label=features)
    plt.title(f'Кластер {cluster_id} | Важность признаков')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

```python
months = year_data['timestamp'].dt.month.unique()
monthly_data = {}

for month in months:
    # Фильтрация данных за текущий месяц
    month_data = year_data[year_data['timestamp'].dt.month == month].copy()

    # Проверяем, есть ли столбец Cluster
    if 'Cluster' not in month_data.columns:
        print(f"В данных за месяц {month} отсутствует столбец 'Cluster'. Пропускаем.")
        continue

    reconstruction_errors = np.zeros(len(month_data))

    # Получаем числовые признаки для масштабирования
    numeric_columns = month_data.select_dtypes(include=['float64', 'int64']).columns

    # Проходим по каждому кластеру, чтобы обработать данные пакетно
    for cluster_label in month_data['Cluster'].unique():
        cluster_mask = month_data['Cluster'] == cluster_label
        cluster_data = month_data.loc[cluster_mask].copy()

        # Проверяем, что для кластера есть модель и scaler
        if cluster_label not in autoencoders:
            print(f"Модель для кластера {cluster_label} отсутствует. Пропускаем данные этого кластера.")
            continue

        scaler = autoencoders[cluster_label]['scaler']
        ae_model = autoencoders[cluster_label]['model']
        features = autoencoders[cluster_label]['features']

        # Проверяем наличие всех признаков в данных, если нет — добавляем с 0
        missing_cols = [col for col in features if col not in cluster_data.columns]
        if missing_cols:
            for col in missing_cols:
                cluster_data[col] = 0.0

        # Отбираем только необходимые признаки в правильном порядке
        cluster_data_features = cluster_data[features]

        # Масштабируем
        X_scaled = scaler.transform(cluster_data_features)

        # Предсказываем реконструкцию
        X_reconstructed = ae_model.predict(X_scaled)

        # Вычисляем ошибку реконструкции (MSE по признакам)
        errors = np.mean(np.square(X_scaled - X_reconstructed), axis=1)

        # Записываем ошибки обратно в общий массив
        reconstruction_errors[cluster_mask] = errors

    # Добавляем столбец с ошибками
    month_data['ReconstructionError'] = reconstruction_errors

    # Сохраняем обработанный месяц
    monthly_data[month] = month_data

    # Сохраняем в CSV
    month_data.to_csv(f'month_data_{month}.csv', index=False)

# Объединяем все месяцы в один DataFrame
all_months_data = pd.concat(monthly_data.values(), axis=0).reset_index(drop=True)
all_months_data.to_csv('all_months_data.csv', index=False)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def safe_analyze_errors(errors_series):
    """Безопасный анализ ошибок с защитой от переполнения"""
    # 1. Фильтрация и преобразование ошибок
    errors = errors_series[np.isfinite(errors_series)]
    log_errors = np.log10(errors + 1e-12)  # Защита от нулевых значений
    
    # 2. Расчет статистики в логарифмической шкале
    percentiles = [0.25, 0.5, 0.75, 0.9, 0.95, 0.99]
    stats = {
        'original': errors.describe(percentiles=percentiles),
        'log': pd.Series(log_errors).describe(percentiles=percentiles)
    }
    
    # 3. Определение разумных границ визуализации
    log_min = stats['log']['min']
    log_max = min(stats['log']['max'], log_min + 15)  # Ограничиваем диапазон
    
    # 4. Визуализация
    plt.figure(figsize=(15, 6))
    
    # График 1: Логарифмированное распределение
    plt.subplot(1, 2, 1)
    bins = np.linspace(log_min, log_max, 50)
    sns.histplot(log_errors, bins=bins, kde=True)
    
    # Добавляем линии перцентилей
    percentiles_to_show = [0.5, 0.75, 0.95, 0.99]
    colors = ['green', 'blue', 'orange', 'red']
    
    for p, color in zip(percentiles_to_show, colors):
        val = stats['log'][f'{int(p*100)}%']
        plt.axvline(val, color=color, linestyle='--', 
                   label=f'{int(p*100)}%: {10**val:.1e}')
    
    plt.title('Логарифм ошибок реконструкции')
    plt.xlabel('log10(Ошибка)')
    plt.ylabel('Частота')
    plt.legend()
    
    # График 2: Основная масса данных (99% перцентиль)
    plt.subplot(1, 2, 2)
    main_data = errors[errors <= 10**stats['log']['99%']]
    sns.histplot(main_data, bins=50, kde=True)
    plt.title('Распределение 99% ошибок')
    plt.xlabel('Ошибка реконструкции')
    
    plt.tight_layout()
    plt.show()
    
    # 5. Вывод статистики
    print("="*60)
    print("Статистика ошибок (оригинальные значения):")
    print(stats['original'])
    print("\nСтатистика в логарифмической шкале:")
    print(stats['log'])
    print("="*60)
    
    # 6. Рекомендуемые пороги
    thresholds = {
        'normal': 10**stats['log']['75%'],
        'warning': 10**stats['log']['95%'],
        'critical': 10**stats['log']['99%']
    }
    
    print("\nРекомендуемые пороги для мониторинга:")
    for level, val in thresholds.items():
        count = (errors > val).sum()
        print(f"- {level.upper()}: > {val:.1f} ({count} примеров, {count/len(errors)*100:.2f}%)")
    
    return thresholds

# Применение функции
thresholds = safe_analyze_errors(all_months_data['ReconstructionError'])
```

```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output, State
import plotly.graph_objs as go
import pandas as pd
import numpy as np
from datetime import timedelta
import random
from dash.exceptions import PreventUpdate
from dash import ctx

# Пример загрузки данных
all_months_data['timestamp'] = pd.to_datetime(all_months_data['timestamp'])
all_months_data = all_months_data.sort_values('timestamp')

external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']
app = dash.Dash(__name__, external_stylesheets=external_stylesheets)
server = app.server

# Получаем начальные и конечные даты
date_min = all_months_data['timestamp'].min()
date_max = all_months_data['timestamp'].max()

# Выбираем 3 случайных показателя
sensor_columns = [col for col in all_months_data.columns if col not in ['timestamp', 'Cluster', 'ReconstructionError', 'Валидатор ошибки', 'Ошибка: FMI', 'Ошибка: SPN']]
def random_sensors():
    return random.sample(sensor_columns, 3)

app.layout = html.Div([
    dcc.Location(id='url', refresh=False),
    html.H1("Anomaly Detection Viewer"),

    dcc.DatePickerRange(
        id='date-picker',
        start_date=date_max - timedelta(days=7),
        end_date=date_max,
        display_format='YYYY-MM-DD',
        min_date_allowed=date_min,
        max_date_allowed=date_max
    ),

    dcc.Graph(id='error-graph'),
    html.Div(id='anomaly-info', style={'marginTop': '20px'}),
    html.Div(id='anomaly-detail-layout')
])

@app.callback(
    Output('error-graph', 'figure'),
    Input('date-picker', 'start_date'),
    Input('date-picker', 'end_date')
)
def update_graph(start_date, end_date):
    df = all_months_data[(all_months_data['timestamp'] >= start_date) & (all_months_data['timestamp'] <= end_date)]
    fig = go.Figure()
    fig.add_trace(go.Scatter(
        x=df['timestamp'],
        y=df['ReconstructionError'],
        mode='lines+markers',
        name='Reconstruction Error',
        marker=dict(color='blue'),
        hoverinfo='text',
        customdata=df.index,
        hovertext=[f"Click to view Anomaly {i}" for i in df.index]
    ))
    fig.update_layout(title="Anomaly Detection Over Time", xaxis_title="Timestamp", yaxis_title="Reconstruction Error")
    return fig

@app.callback(
    Output('anomaly-info', 'children'),
    Input('error-graph', 'clickData')
)
def display_anomaly_info(clickData):
    if clickData is None:
        return html.Div("Click on a point in the graph to view anomaly details.")

    index = clickData['points'][0]['pointIndex']
    row = all_months_data.iloc[index]
    sensors = random_sensors()

    return html.Div([
        html.H3(f"Anomaly {index}"),
        html.P(f"Cluster: {row['Cluster']}"),
        html.P(f"Reconstruction Error: {row['ReconstructionError']:.4f}"),
        html.H4("Anomaly Values"),
        html.Div([
            html.Div(f"{col}: {row[col]}", style={'backgroundColor': 'white', 'padding': '5px', 'marginBottom': '2px'}) for col in sensors
        ]),
        html.Br(),
        dcc.Link('View detailed anomaly chart', href=f'/anomaly/{index}', refresh=False)
    ])

@app.callback(
    Output('anomaly-detail-layout', 'children'),
    Input('url', 'pathname')
)
def render_anomaly_detail(pathname):
    if not pathname.startswith('/anomaly/'):
        raise PreventUpdate

    index = int(pathname.split('/')[-1])
    center_time = all_months_data.iloc[index]['timestamp']
    start_time = center_time - timedelta(minutes=30)
    end_time = center_time + timedelta(minutes=30)
    df = all_months_data[(all_months_data['timestamp'] >= start_time) & (all_months_data['timestamp'] <= end_time)]
    default_sensors = random_sensors()

    return html.Div([
        html.H2(f"Detailed Chart for Anomaly {index}"),
        dcc.Checklist(
            id='sensor-selector',
            options=[{'label': col, 'value': col} for col in sensor_columns],
            value=default_sensors,
            labelStyle={'display': 'block'},
            style={'overflowY': 'scroll', 'height': '200px'}
        ),
        dcc.Graph(id='sensor-graph'),
        dcc.Store(id='detail-index', data=index)
    ])

@app.callback(
    Output('sensor-graph', 'figure'),
    Input('sensor-selector', 'value'),
    State('detail-index', 'data')
)
def update_sensor_graph(selected_sensors, index):
    center_time = all_months_data.iloc[index]['timestamp']
    start_time = center_time - timedelta(minutes=30)
    end_time = center_time + timedelta(minutes=30)
    df = all_months_data[(all_months_data['timestamp'] >= start_time) & (all_months_data['timestamp'] <= end_time)]

    fig = go.Figure()
    for sensor in selected_sensors:
        fig.add_trace(go.Scatter(
            x=df['timestamp'],
            y=df[sensor],
            mode='lines',
            name=sensor
        ))
    fig.update_layout(title="Sensor Values Around Anomaly", xaxis_title="Timestamp", yaxis_title="Value")
    return fig

if __name__ == '__main__':
    app.run(debug=True)
```
